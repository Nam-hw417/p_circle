```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 불러오기
df = pd.read_csv("titanic.csv")

# 성별에 따른 생존 여부 비율 계산
sex_survived = df.groupby(['Sex', 'Survived'])['Survived'].count()
sex_total = df.groupby('Sex')['PassengerId'].count()
sex_survival_rate = sex_survived / sex_total * 100

# 시각화
fig, ax = plt.subplots()
sex_survival_rate.unstack().plot(kind='bar', ax=ax)
ax.set(title='Survival Rate by Gender', ylabel='Survival Rate (%)', xlabel='Gender')
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 불러오기
# df = pd.read_csv("titanic.csv")

# 생존 여부에 따른 나이 분포 계산
survived_age = df.groupby(['Survived'])['Age']

# 시각화
fig, ax = plt.subplots()
survived_age.plot(kind='hist', alpha=0.5, bins=20, ax=ax)
ax.set(title='Survival Distribution by Age', ylabel='Count', xlabel='Age')
ax.legend(['Did not survive', 'Survived'])
plt.show()
```

```python
df = pd.read_csv('train (1).csv')

# 결측치 처리
df['Age'] = df['Age'].fillna(df['Age'].mean())  # Age 열의 결측치를 평균값으로 대체
df = df.drop(['Cabin', 'Embarked'], axis=1)     # Cabin과 Embarked 열 삭제

# 범주형 변수 처리
df['Sex'] = df['Sex'].replace(['male', 'female'], [0, 1])      # Sex 열의 값을 0과 1로 변환
df['Age_cat'] = pd.cut(df['Age'], bins=[0, 9, 19, 29, 39, 49, 59, 69, 100], labels=[0, 1, 2, 3, 4, 5, 6, 7]) # Age 열의 값을 8개의 구간으로 나누어 Age_cat 열에 저장
df = df.drop(['Name', 'Ticket', 'Age'], axis=1)               # Name, Ticket, Age 열 삭제

# 더미 변수 생성
df = pd.get_dummies(df, columns=['Pclass', 'Age_cat'], prefix=['Pclass', 'Age_cat'])

# 최종 전처리된 데이터셋 출력
print(df.head())
```

```python
df.head()
```

```python
df = df.drop('PassengerId', axis = 1)
```

```python
y = df['Survived']
X = df.drop(['Survived'], axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
import re

# 호칭(Title) 정보를 추출하는 함수를 정의합니다.
def get_title(name):
    title_search = re.search(' ([A-Za-z]+)\.', name)
    if title_search:
        return title_search.group(1)
    return ""

# Name 변수에서 성별(Sex)과 호칭(Title) 정보를 추출하여 새로운 변수를 생성합니다.
titanic['Sex'] = titanic['Sex'].map({'male': 0, 'female': 1}) # 성별을 0(male)과 1(female)로 변환합니다.
titanic['Title'] = titanic['Name'].apply(get_title)

# 호칭(Title) 정보를 기반으로 일부 호칭을 그룹화합니다.
titanic['Title'] = titanic['Title'].replace(['Lady', 'Countess', 'Capt', 'Col',\
                                             'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Other')
titanic['Title'] = titanic['Title'].replace('Mlle', 'Miss')
titanic['Title'] = titanic['Title'].replace('Ms', 'Miss')
titanic['Title'] = titanic['Title'].replace('Mme', 'Mrs')
```
